```python
#Print the top 5 customers (or stores) by total sales amount.
# Create total sales column
df_sales['TOTAL SALES'] = df_sales['RETAIL SALES'] + df_sales['WAREHOUSE SALES']

# Get top 5 suppliers
top_5_suppliers = (
    df_sales.groupby('SUPPLIER')['TOTAL SALES']
      .sum()
      .sort_values(ascending=False)
      .head(5)
)

print(top_5_suppliers)
```
