```python
# Import necessary libraries
import pandas as pd
import numpy as np
```

```python
#Read the dataset
from google.colab import drive
drive.mount('/content/drive')

df_sales = pd.read_csv("/content/Warehouse_and_Retail_Sales.csv")
```

```python
df_sales.shape
```

```python
df_sales.head()
```

```python
df_sales.isna().sum()
```

```python
print("Missing values per column:")
print(df_sales.isnull().sum())
```

```python
# SUPPLIER - 167 missing (167/307645 = 0.054%),
# Decision: fill with 'Unknown' as it's categorical
df_sales['SUPPLIER'].fillna('Unknown', inplace=True)

# ITEM TYPE - 1 missing (1/307645 = 0.0003%),
# Decision: fill with mode
df_sales['ITEM TYPE'].fillna(df_sales['ITEM TYPE'].mode()[0], inplace=True)

# RETAIL SALES - 3 missing (3/307645 = 0.000975%),
# Decision: fill with Median
df_sales['RETAIL SALES'] = pd.to_numeric(df_sales['RETAIL SALES'], errors='coerce')
df_sales['RETAIL SALES'].fillna(df_sales['RETAIL SALES'].median(), inplace=True)

print(f"\nOriginal shape: {df_sales.shape}")
print(f"Cleaned shape: {df_sales.shape}")
print(f"\nRemaining missing values:")
print(df_sales.isnull().sum())
```

# **FEATURE** **ENGINEERING**

```python
# CREATE A NEW COLUMN 'DATE' AND MAKE IT A DATETIME DATATYPE
df_sales['DATE'] = df_sales['YEAR'].astype(str) + '-' + df_sales['MONTH'].astype(str)
df_sales['DATE'] = pd.to_datetime(df_sales['DATE'])
df_sales.info()
```

```python
# CREATE TOTAL SALES COLUMN TO SEE TOTAL SALES FROM EACH TRANSACTIONS
df_sales['TOTAL SALES'] = df_sales['RETAIL SALES'] + df_sales['WAREHOUSE SALES']
df_sales.head()
```
